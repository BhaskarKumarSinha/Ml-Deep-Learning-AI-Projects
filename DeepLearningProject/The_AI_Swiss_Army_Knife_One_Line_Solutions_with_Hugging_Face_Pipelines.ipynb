```python
from diffusers import StableDiffusionPipeline
import torch

# Choose a diffusion model from Hugging Face Hub
# I'll use the Stable Diffusion v1.5 model
model_id = "runwayml/stable-diffusion-v1-5"

# Load the pipeline
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)

# Move the pipeline to GPU if available
if torch.cuda.is_available():
    pipe = pipe.to("cuda")
```

```python
# Generate an image
prompt = "a photo of an astronaut riding a horse on the moon"
image = pipe(prompt).images[0]

# You can add more prompts and generate multiple images here
# prompt2 = "a digital painting of a futuristic city"
# image2 = pipe(prompt2).images[0]
```

# Task
Generate multiple images using the "stabilityai/stable-diffusion-2-1" model with different prompts and parameters. For each generated image, create a caption using the "Salesforce/blip-image-captioning-base" model and display the image with its corresponding caption.

## Generate multiple images

### Subtask:
Modify the existing code to generate several images with different prompts.


**Reasoning**:
The subtask requires generating multiple images with different prompts using a different model. This can be achieved by modifying the existing code to use the new model ID, create a list of prompts, and iterate through the prompts to generate and store the images.



```python
from diffusers import StableDiffusionPipeline
import torch

# Choose a diffusion model from Hugging Face Hub
# Use the Stable Diffusion 2.1 model
model_id = "stabilityai/stable-diffusion-2-1"

# Load the pipeline
# Use torch_dtype=torch.float16 if you have a compatible GPU for faster inference
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)

# Move the pipeline to GPU if available
if torch.cuda.is_available():
    pipe = pipe.to("cuda")

# Create a list of prompts
prompts = [
    "a man standing on a mountain",
    "a futuristic cityscape",
    "a watercolor painting of a forest"
]

# Store generated images
generated_images = []

# Generate images for each prompt
for prompt in prompts:
    image = pipe(prompt).images[0]
    generated_images.append(image)
```

```python
for i, image in enumerate(generated_images):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

## Caption each image

### Subtask:
Iterate through the generated images and use the image-to-text pipeline to generate a caption for each one.


**Reasoning**:
Iterate through the generated images and use the image-to-text pipeline to generate a caption for each one, storing the captions in a list.



```python
generated_captions = []
for img in generated_images:
    caption_results = image_captioner(img)
    generated_captions.append(caption_results[0]["generated_text"])

print("Generated Captions:", generated_captions)
```

## Display images with captions

### Subtask:
Display each image along with its generated caption.


**Reasoning**:
Iterate through the generated images and captions, display each image with its caption as the title.



```python
for i, image in enumerate(generated_images):
    plt.imshow(image)
    plt.axis('off')
    plt.title(generated_captions[i], fontsize=12, color="blue")
    plt.show()
```
